```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams["figure.figsize"] = (8,6)
```

```python
pd.crosstab(df["partyabbre"].map({"JNP":"Janata Party","JD": "Janata Dal","IND": "Independent", "INC":"Indian National Congress","INC(I)":"Indian National Congress", "BJP": "Bharatiya Janata Party" }), df["st_name"], margins=True)
```

```python
sns.set_theme(style="darkgrid")
plt.figure(figsize = (150,130))
sns.countplot(x="year", hue = "st_name", data=df)
plt.yticks(size=70)
plt.xticks(size=70)  
plt.legend(fontsize= 70)
           
```
